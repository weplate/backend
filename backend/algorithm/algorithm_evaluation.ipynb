```python
import pandas as pd
import numpy as np
```

```python
sdf = pd.read_csv(r'C:/Users/Penelope/Documents/GitHub/Weplate/backend/backend/algorithm/result/student_profile/nutrient_limit_male.csv')
df = pd.read_csv(r'C:\Users\Penelope\Documents\GitHub\Weplate\backend\backend\algorithm\female_may_lsqp.csv')
df['protein'] = df['protein'].astype(float)
df['carbohydrate'] = df['carbohydrate'].astype(float)
df['calories'] = df['calories'].astype(float)
df['total_fat'] = df['total_fat'].astype(float)
df['saturated_fat'] = df['saturated_fat'].astype(float)
```

```python
df.drop(index=0, inplace=True)
df = df.reset_index(drop=True)
df.head(3)
```

```python
def check_exceed(nutrients, nutrient_lims):
    d = []
    for ind_n, nutrient_lims in enumerate(nutrient_lims):
        if nutrients[ind_n] > nutrient_lims[0] and nutrients[ind_n] < nutrient_lims[1]:
            d.append(0)
        else:
            d.append(1)
    d.append(sum(d))
    return np.array(d)
```

```python
nutrient_check = np.zeros(6)
for df_index in range(len(df)):
    protein_lim, carbohydrate_lim, calories_lim, total_fat_lim, saturated_fat_lim = sdf.loc[df.loc[df_index, 'student_id'], ['protein_lim', 'carbohydrate_lim','calories_lim','total_fat_lim','saturated_fat_lim']]
    protein, carbohydrate, calories, total_fat, saturated_fat = df.loc[df_index, ['protein', 'carbohydrate', 'calories', 'total_fat', 'saturated_fat']]
    limits = [eval(protein_lim), eval(carbohydrate_lim), eval(calories_lim), eval(total_fat_lim), eval(saturated_fat_lim)]
    nutrients = [protein, carbohydrate, calories, total_fat, saturated_fat]
    exceed = check_exceed(nutrients, limits)
    nutrient_check = np.vstack((nutrient_check, exceed))


```

```python
edf = pd.DataFrame(nutrient_check, columns=['protein_exceed', 'carbohydrate_exceed', 'calories_exceed', 'total_fat_exceed', 'saturated_fat_exceed', 'num_exceed'])
```
